```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, "../../")
from fromscratch.unsupervised.kmeans import KMeans
```

### Simple toy dataset

```python
n1, n2, n3 = 300, 450, 120

X = np.vstack((
    np.random.multivariate_normal([0, 0], np.diag([0.2, 0.2]), n1),
    np.random.multivariate_normal([3, 1], np.diag([0.2, 0.2]), n2),
    np.random.multivariate_normal([2, 2], np.diag([0.1, 0.1]), n3)
))

y = np.array([1]*n1 + [2]*n2 + [3]*n3)  # underlying/unknown data clustering
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5, s=40)
plt.show()
```

### K-Means Clustering

#### K = 3 clusters

```python
kmeans = KMeans(K=3)
```

```python
kmeans.fit(X)
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.assigned_means, alpha=0.5, s=40)
plt.show()
```

#### K = 2 clusters

```python
kmeans = KMeans(K=2)
```

```python
kmeans.fit(X)
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.assigned_means, alpha=0.5, s=40)
plt.show()
```
